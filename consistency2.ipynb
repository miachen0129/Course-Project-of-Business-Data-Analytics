# 分数转换

## 归一化

```python
def minmax(data):
    """
    对df进行min-max归一化
    :param data: （每个组的)dataframe
    :return Senti部分经过归一化
    """
    tags = data.Tag.unique()
    for tag in tags:
        df = data[data['Tag']==tag]
        S_df = df.loc[:,'S11':'S34']
        # S_minmax = S_df.apply(lambda x: (x - x.min())/(x.max()-x.min()))
        # df[cols2] = S_minmax
        df = scoreTransform(df)
        data[data['Tag']==tag] = df
    return data
    
def scoreTransform(df):
    """
    将df转化至0-25区间
    :param df: 经过minmax以后的df
    :return 经过重新计算分数后的df
    """
    for i in range(1,4):
        # 三位评委i=1，2，3
        X_cols = ['X%s%s'%(i,j) for j in range(1,5)] # 打分X的列名
        S_cols = ['S%s%s'%(i,j) for j in range(1,5)] # 情感分S的列名
        for i,S in enumerate(S_cols):
            Sij = df[[S]] # Sij列，为dataframe
            
            Xij = df.loc[:,X_cols[i]] # Xij列
            maxX,minX = Xij.max(), Xij.min()
            Sij = Sij.apply(lambda x: x* (maxX-minX) +minX)
            df[[S]] = Sij
    return df
```

## 均分计算

```python
minmax_df = minmax(df)

for i in range(1,4):
    col_names = ["S%s%s"%(i,j) for j in range(1,5)]
    minmax_df["S%s"%i] = minmax_df[col_names].sum(axis=1)
    
for j in range(1,5):
    col_names = ["S%s%s"%(i,j) for i in range(1,4)]
    minmax_df["AvgS%s"%j] =minmax_df[col_names].mean(axis=1)
    
minmax_df['AvgSenti'] = minmax_df[['S1','S2','S3']].mean(axis=1)
```

## kendall

### 按专家分

```python
tags = df.Tag.unique()

from scipy.stats import kendalltau
def kendall_by_judge(grouped_df):
    result_dict = {}
    k_ls = [[], [], []]
    p_ls = [[], [], []]
    for name,group in grouped_df:
        for i in range(1,4):
            col1, col2 = 'X%s'%i, 'S%s'%i
            cal_df = group[[col1,col2]]
            cal_df.dropna(how='any', axis=0, inplace=True)
            k,p = kendalltau(cal_df[col1],cal_df[col2])
            k_ls[i-1].append(k)
            p_ls[i-1].append(p)
    for i in range(1,4):
        result_dict['Kendall_Judge%s'%i] = k_ls[i-1]
        result_dict['p_Judge%s'%i] = p_ls[i-1]
    return result_dict
```

```python
judge_kendall = pd.DataFrame(kendall_by_judge(grouped_minmax),index = tags)
```

```python
def spearman_by_judge(grouped_df):
    result_dict = {}
    k_ls = [[], [], []]
    p_ls = [[], [], []]
    for name,group in grouped_df:
        for i in range(1,4):
            col1, col2 = 'X%s'%i, 'S%s'%i
            cal_df = group[[col1,col2]]
            cal_df.dropna(how='any', axis=0, inplace=True)
            k,p = spearmanr(cal_df)
            k_ls[i-1].append(k)
            p_ls[i-1].append(p)
    for i in range(1,4):
        result_dict['Spearm_Judge%s'%i] = k_ls[i-1]
        result_dict['p_Judge%s'%i] = p_ls[i-1]
    return result_dict
```

### 按维度+学科分

```python
def spearman_by_dim(df):
    Sp_ls = []
    p_ls = []
    for i in range(1,5):
        XnS = df[['Avg%s'%i, 'AvgS%s'%i]]
        XnS.dropna(how='any', axis=0, inplace=True)
        Spearm, p = spearmanr(XnS)
        Sp_ls.append(Spearm)
        p_ls.append(p)
    return Sp_ls, p_ls

S_ls = [[],[],[],[]]
p_ls = [[],[],[],[]]
for name, group in grouped_minmax:
    Sp, p = spearman_by_dim(group)
    for i in range(4):
        S_ls[i].append(Sp[i])
        p_ls[i].append(p[i])
        
result_dict = {}
for i in range(1,5):
    result_dict['Spearm_Dim%s'%i] = S_ls[i-1]
    result_dict['p_Dim%s'%i] = p_ls[i-1]
spearman_dim = pd.DataFrame(result_dict, index = tags)
```
